```python
import tensorflow as tf
import numpy as np
import input_data
```

```python
mnist_width = 28
n_visible = mnist_width * mnist_width
n_hidden = 10
nBatch = 128
```

```python
# create node for input data
X = tf.placeholder("float", [nBatch, n_visible], name='X')
H = tf.placeholder("float", [nBatch, n_hidden], name='H')
```

```python

# create nodes for hidden variables
C_max = 4 * np.sqrt(6. / (n_visible + n_hidden))
C_init = tf.random_uniform(shape=[n_visible, n_hidden],
                           minval=-C_max,
                           maxval=C_max)

C = tf.Variable(C_init, name='C')
```

```python
print C.get_shape()
print X.get_shape()
print H.get_shape()
```

```python
C2 = tf.expand_dims(tf.transpose(C),0)
X2 = tf.expand_dims(X,1)
print C2.get_shape()
print X2.get_shape()
```

```python
dist = tf.reduce_sum(tf.square(tf.sub(X2,C2)),2)
cost = tf.reduce_sum(tf.reduce_min(dist,1))
#print assign.get_shape()
```

```python
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(cost)  # construct an optimizer
#train_op = tf.train.AdamOptimizer().minimize(cost)
mnist = input_data.read_data_sets("/mnt/hdd1/thkim/dataset/MNIST", one_hot=True)
trX, trY, teX, teY = mnist.train.images, mnist.train.labels, mnist.test.images, mnist.test.labels
```

```python
# Launch the graph in a session
sess = tf.Session()

# you need to initialize all variables
tf.initialize_all_variables().run(session=sess)
print('-1', sess.run(cost, feed_dict={X: teX[:128]}))
for i in range(100):
    for start, end in zip(range(0, len(trX), 128), range(128, len(trX), 128)):
        input_ = trX[start:end]
        sess.run(train_op, feed_dict={X: input_})
    print(i, sess.run(cost, feed_dict={X: teX[:128]}))

```
